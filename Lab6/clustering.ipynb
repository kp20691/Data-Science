```python
#training data
x = df.iloc[:,1:10].values

#country label 
y = df.iloc[:,0]
```

```python
from sklearn import preprocessing 
df_stsc = preprocessing.scale(x)
print(df_stsc)
```

```python
#Plotting Elbow Curve
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
from sklearn import metrics

model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,10))
visualizer.fit(df_stsc)    
visualizer.poof()
```

```python
model=KMeans(n_clusters=3)
model.fit(df_stsc)
```

```python
print(model.labels_)
```

```python
pd.Series(model.labels_).value_counts()
```

```python
df['preds'] = model.labels_
df
```

```python
df[df.preds == 0]
```

```python
df[df.preds == 1]
```

```python
df[df.preds==2]
```

```python
#Visualize clusters: Feature Pair-2
plt.figure(figsize=(12,6))
plt.scatter(df_stsc[:,4],df_stsc[:,8],c=df.preds) # income vs gdpp
plt.title("Income vs GDPP (Visualize KMeans Clusters)", fontsize=15)
plt.xlabel("Income", fontsize=12)
plt.ylabel("GDPP", fontsize=12)
plt.rcParams['axes.facecolor'] = 'lightblue'
plt.show()
```

```python
df.preds = df.preds.replace({1:"Under-developed", 0: "Developing", 2: "Developed"})
df
```

In this assessment, we plotted various features and found out that GDPP, child_mort and income are some of the major features 

We can observe a correlation between income and gdpp, After doing the clustering visualisation, we can observse that developed countires have higher gdpp and income and there is a linear relation between these two features..

The countries which have low gdp have higher child_mort due to less facilities available and bad living standards

Also we applied normalisation to our data before modelling our Kmeans clustering algorithm to remove interpretation of different features weighted, because all features possess equal importance while clustering , thus we applied standard scaling to remove such inconsistencies 

Thus by following the above measures we modelled A Kmeans clustering algorithm with 3 clusters.
